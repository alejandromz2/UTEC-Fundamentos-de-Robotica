```python
#Metodo Gradiente descendiente
import numpy as np
cos=np.cos; sin=np.sin
xd = np.array([-2, -3]) # Desired value in the Cartesian space
q = np.array([-0.5, -1]) # Initial value in the joint space
epsilon = 1e-3
max_iter = 1000 # Maximum number of iterations
alpha = 0.1


# Iterations: Gradient descent
for i in range(max_iter):
    q1 = q[0]; q2 = q[1];
    J = np.array([[-q2*cos(q1)-0.2*sin(q1), -sin(q1)],[ -q2*sin(q1)+0.2*cos(q1), cos(q1)]])
    f = np.array([-q2*sin(q1)+0.2*cos(q1), q2*cos(q1)+0.2*sin(q1)])
    e = xd-f
    q = q + alpha*np.dot(J.T, e)
    print('iteracion:{}, error:{}'.format(i,np.linalg.norm(e)))
# End condition
    if (np.linalg.norm(e) < epsilon):
        break
    if (i==max_iter-1):
        print("El algoritmo diverge, escoger un alpha más pequeño")
print(q)
```
